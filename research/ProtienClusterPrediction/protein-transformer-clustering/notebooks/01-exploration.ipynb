# Exploration of Protein Sequences

This notebook contains exploratory data analysis (EDA) and visualization of protein sequences. The goal is to understand the distribution of sequences, their properties, and prepare for clustering based on resistance patterns.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set visualization style
sns.set(style='whitegrid')

# Load protein sequences from FASTA files
def load_fasta_sequences(fasta_directory):
    sequences = []
    for filename in os.listdir(fasta_directory):
        if filename.endswith('.fasta'):
            with open(os.path.join(fasta_directory, filename)) as f:
                seq = ''
                for line in f:
                    if not line.startswith('>'):
                        seq += line.strip()
                sequences.append(seq)
    return sequences

# Example usage
fasta_dir = '../data/raw'
protein_sequences = load_fasta_sequences(fasta_dir)
print(f'Loaded {len(protein_sequences)} protein sequences.')
```

## Sequence Length Distribution

Let's analyze the distribution of protein sequence lengths.

```python
sequence_lengths = [len(seq) for seq in protein_sequences]

plt.figure(figsize=(10, 6))
sns.histplot(sequence_lengths, bins=30, kde=True)
plt.title('Distribution of Protein Sequence Lengths')
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.show()
```

## GC Content Analysis

Next, we will calculate the GC content of the sequences and visualize its distribution.

```python
def calculate_gc_content(sequence):
    g_count = sequence.count('G')
    c_count = sequence.count('C')
    return (g_count + c_count) / len(sequence) if len(sequence) > 0 else 0

gc_contents = [calculate_gc_content(seq) for seq in protein_sequences]

plt.figure(figsize=(10, 6))
sns.histplot(gc_contents, bins=30, kde=True)
plt.title('Distribution of GC Content in Protein Sequences')
plt.xlabel('GC Content')
plt.ylabel('Frequency')
plt.show()
```

## Summary Statistics

Let's summarize the statistics of the protein sequences.

```python
summary_df = pd.DataFrame({'Length': sequence_lengths, 'GC Content': gc_contents})
summary_stats = summary_df.describe()
summary_stats
```
